## Las variables
1. CP --> Costo de producto
2. CD --> Cantidad de dinero
3. Cambio

## Formulas
$Cambio = CD - CP$

## Diagrama de flujo LucidChart
<img src="https://raw.githubusercontent.com/JazminAguirre128/Calcular-cambio-if-else-/refs/heads/main/Df%20if%20else%20cambio%202.png" width="350" height="500" alt="Texto descriptivo">


```python
# Para cuando es verdadero (no)
# Calculando el cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# Si es verdadero mostrar si no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CP >= CD:
  print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
print("Fin del programa")
print("Meredith Aguirre NL = 2")
```
